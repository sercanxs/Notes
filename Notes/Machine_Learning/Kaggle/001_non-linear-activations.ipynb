```python
import torch
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
# Create a tensor
A = torch.arange(-10, 10, 1, dtype=torch.float32)
A, A.dtype
```

```python
plt.plot(A, color="darkblue")
plt.xlabel('index')
plt.ylabel('values')
plt.show()
```



**ReLU** --> https://pytorch.org/docs/main/generated/torch.nn.ReLU.html#torch.nn.ReLU

```python
def relu(x: torch.Tensor) -> torch.Tensor:                                 
  return torch.maximum(torch.tensor(0), x)
```

```python
relu(A)
```

```python
plt.plot(relu(A), color="darkblue")
plt.title('ReLU Function')
plt.xlabel('index')
plt.ylabel('relu(x)')
plt.show()
```


**Sigmoid** --> https://pytorch.org/docs/main/generated/torch.nn.Sigmoid.html#torch.nn.Sigmoid
<br>

```python
def sigmoid(x):
  return 1 / (1 + torch.exp(-x))
```

```python
sigmoid(A)
```

```python
plt.plot(sigmoid(A), color="darkblue")
plt.title('Sigmoid Function')
plt.xlabel('index')
plt.ylabel('sigmoid(x)')
plt.show()
```


**Softmax** --> https://pytorch.org/docs/main/generated/torch.nn.Softmax.html#torch.nn.Softmax

```python
def softmax(x):
    e_x = torch.exp(x-torch.max(x))
    return e_x/e_x.sum()
```

```python
softmax(A)
```

```python
plt.plot(softmax(A), color="darkblue")
plt.title('Softmax Function')
plt.xlabel('index')
plt.ylabel('softmax(x)')
plt.show()
```

```python
softmax(A).sum().item()
```

```python
round(softmax(A).sum().item())
```
