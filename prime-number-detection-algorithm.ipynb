Now we create an algorithm that captures prime numbers based on interval. Let's search prime numbers between 80 and 87. If we denote left numbers as "a" right numbers minimum value is 3a and maximum value is a^2. so if a is 29 it's minimum value on the right is 87. We search numbers those are less than 87. So we don't start from the a=29 and we need to go top to reach the interval. So our starting point is 27.(28 is even number. We skip that) First column of the a=27 is 81 and this number is in the interval. Maybe there are numbers those are in the interval, in the right. We go to right and reach the 135. 135 is greater than 87 so it is not in the interval. We don't need to go right because numbers are increasing left to right. values in the right are not in the interval. Let's go top. We reach 75 and it is less than 80 so it is not in the interval. Notice that numbers are decreasing from bottom to top. So all numbers in the top are not in the interval. So we should never go to column 1 again. 75 is less than 80. To reach the interval we go to right and reach the 125. 125 is greater than 87 so it is not in the interval. Right numbers are not in the interval so we go top but not column 1. We reach the 115 and it is not in the interval. We go top again and reach the 105. We follow this direction. --> 95,85,119,75,105,91,77,99,81. When we reach the 81 there are no values in the top and in the right. So we stop here. When we reach the numbers those are in the interval we find non-prime numbers in the interval. So if we eliminate non-prime numbers and even numbers we find prime numbers. In this example 81,85 and all even values 82,84,86 are non prime numbers so 83 is prime number. Notice that number 2 is the only even prime number. So if interval contains 2 we can add 2 as a prime number.

<h3>Let's write code</h3>

```python
import math




primeNumberLeft=1     # We search prime numbers greater than this
primeNumberRight=100    # We search prime numbers less than this

if(primeNumberLeft<1):
    raise Exception("primeNumberLeft must be greater than 0. Please change it")
if(primeNumberLeft>=primeNumberRight):
    raise Exception("primeNumberLeft must be less than primeNumberRight. Please change it")
        


DetectionIndex=(primeNumberRight-3)/6 # We try to find which index's minimum value is right value of the interval


print(f"Starting index should less than: {DetectionIndex}")
tempVar=DetectionIndex # we create temporary variable

#We search numbers those are not in DetectionIndex. DetectionIndex is the right border of the interval 
#so we take the max index less than DetectionIndex

if(tempVar%1==0): # if DetectionIndex is integer we subtract 1 from DetectionIndex
    #print(tempVar-1)
    SearchBeginningRow=int(tempVar-1)
else:             # if DetectionIndex has a fraction we round it to floor
    #print(math.floor(tempVar))
    SearchBeginningRow=math.floor(tempVar)

print(f"The Starting index is: {SearchBeginningRow}")



SearchColumn=0 # This is the starting column

  

NotPrimeNumbers=[] # This is the list of odd non-prime numbers.

for tempSearchIndex in range(SearchBeginningRow):
    print('Search Index is ='+str(SearchBeginningRow-tempSearchIndex))

    for index in range(SearchColumn,SearchBeginningRow-tempSearchIndex):
         
         x=(SearchBeginningRow-tempSearchIndex)         # This is which row we are in
         NotPrimeNumber=3*(x*(2)+1)+(index)*(2*(x*(2)+1)) # This is the formula of cell value based on row and column

         if(NotPrimeNumber<=primeNumberLeft): # The number is less than interval values. So we go right. 
             SearchColumn+=1                  # When go to top, in the same column there is no value that is in the interval. 
             continue                         # So we never go to same column when we move to top.
         elif(primeNumberLeft<NotPrimeNumber & NotPrimeNumber<primeNumberRight): # If the number is in the interval, we find non-prime
             NotPrimeNumbers.append(NotPrimeNumber)                              # number and we go to right until we go beyond interval
             continue    
         else:                                                                   #If we exceed interval, there is no reason to stay in
             break                                                               # this row. So let's move to top.
         
    if(SearchColumn>=(SearchBeginningRow-tempSearchIndex-1)):  # If there are no values in the top and in the right we can break the
        break                                                  # loop
        


#print(NotPrimeNumbers)
NotPrimeNumbers=set(NotPrimeNumbers) # We eliminate duplicate values
#print(NotPrimeNumbers)

if(primeNumberLeft%2==0):                                               # We eliminate even values
    searchInterval=list(range(primeNumberLeft+1,primeNumberRight,2))
else:
    searchInterval=list(range(primeNumberLeft+2,primeNumberRight,2))


PrimeNumbers=[]


for number in searchInterval:
    if(number in NotPrimeNumbers):   # We eliminate non-prime odd numbers in the interval. 
        continue
    else:
        PrimeNumbers.append(number)
        
if(primeNumberLeft==1):              # If interval contains 2. We add 2 to list because 2 is the only even prime number. 
    PrimeNumbers.append(2)
    print("\nNumber 2 can be found in the last index in below")

print(f"\nPrime Numbers in interval({primeNumberLeft},{primeNumberRight}) are:")
print(PrimeNumbers)
    



```

<h3>Let's create a function</h3>
    

```python
import math



def primeNumberFunc(primeNumberLeft=1,primeNumberRight=100, showIndex=True):    
    
    if(primeNumberLeft<1):
        raise Exception("primeNumberLeft must be greater than 0. Please change it")
    if(primeNumberLeft>=primeNumberRight):
        raise Exception("primeNumberLeft must be less than primeNumberRight. Please change it")
            
    
    
    DetectionIndex=(primeNumberRight-3)/6 
    
    
    print(f"Starting index should less than: {DetectionIndex}")
    tempVar=DetectionIndex 
    
    
    
    if(tempVar%1==0): 
        #print(tempVar-1)
        SearchBeginningRow=int(tempVar-1)
    else:            
        #print(math.floor(tempVar))
        SearchBeginningRow=math.floor(tempVar)
    
    print(f"The Starting index is: {SearchBeginningRow}")
    
    
    
    SearchColumn=0 
    
      
    
    NotPrimeNumbers=[] 
    
    for tempSearchIndex in range(SearchBeginningRow):
        if(showIndex==True):
            print('Search Index is ='+str(SearchBeginningRow-tempSearchIndex))
    
        for index in range(SearchColumn,SearchBeginningRow-tempSearchIndex):
             
             x=(SearchBeginningRow-tempSearchIndex)         
             NotPrimeNumber=3*(x*(2)+1)+(index)*(2*(x*(2)+1)) 
    
             if(NotPrimeNumber<=primeNumberLeft):  
                 SearchColumn+=1                  
                 continue                         
             elif(primeNumberLeft<NotPrimeNumber & NotPrimeNumber<primeNumberRight): 
                 NotPrimeNumbers.append(NotPrimeNumber)                              
                 continue    
             else:                                                                   
                 break                                                               
             
        if(SearchColumn>=(SearchBeginningRow-tempSearchIndex-1)): 
            break                                                  
            
    
    
    #print(NotPrimeNumbers)
    NotPrimeNumbers=set(NotPrimeNumbers) 
    #print(NotPrimeNumbers)
    
    if(primeNumberLeft%2==0):                                               
        searchInterval=list(range(primeNumberLeft+1,primeNumberRight,2))
    else:
        searchInterval=list(range(primeNumberLeft+2,primeNumberRight,2))
    
    
    PrimeNumbers=[]
    
    
    for number in searchInterval:
        if(number in NotPrimeNumbers):  
            continue
        else:
            PrimeNumbers.append(number)
            
    if(primeNumberLeft==1):              
        PrimeNumbers.append(2)
        print("\nNumber 2 can be found in the last index in below")
    
    print(f"\nPrime Numbers in interval({primeNumberLeft},{primeNumberRight}) are:")
    print(PrimeNumbers)
    

```
